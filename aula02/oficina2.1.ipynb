```python
#1 - Importe as bibliotecas pandas, matplotlib, seaborn e use o comando mágico %matplotlib inline para exibir as visualizações no notebook.

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#1 - Carregue os dados em um DataFrame.

source_url = "https://d11s0xq1vqg039.cloudfront.net/videos_conteudo/visual/micro03/oficina/arquivos/dadosoficina03.txt"
data = pd.read_json(source_url)
```

```python
data.head()
```

```python
#2 - Crie uma visualização que mostre o orçamento de todos os filmes do conjunto de dados. Qual o filme com maior orçamento e com menor orçamento?

data_budget = data.sort_values(by="Budget_M", ascending=False).reset_index()
plt.figure(figsize=(10, 10))
plt.title("Orçamento total por filme (milhões)")
g = sns.barplot(x="Budget_M", y="Film", ci = None, data=data_budget)
for index, row in data_budget.iterrows():
    g.text(row.Budget_M, row.name, round(row.Budget_M,2), color='black', ha="left")
plt.show()
```

```python
#3 - Crie uma visualização que mostre a bilheteria de todos os filmes do conjunto de dados. Qual o filme com maior bilheteria e com menor bilheteria?

data_worldwide = data.sort_values(by="Worldwide_Gross_M", ascending=False).reset_index()
plt.figure(figsize=(10, 10))
plt.title("Bilheteria total por filme (milhões)")
g = sns.barplot(x="Worldwide_Gross_M", y="Film", ci = None, data=data_worldwide)
for index, row in data_worldwide.iterrows():
    g.text(row.Worldwide_Gross_M, row.name, round(row.Worldwide_Gross_M,2), color='black', ha="left")
plt.show()
```

```python
#4 - Crie uma visualização que mostre a quantidade de filmes por gênero em cada ano. Em que anos foram feitos menos filmes de ação? E em que ano foram feitos menos filmes de animação?

plt.figure(figsize=(18,5))
plt.title('Filmes por gênero em cada ano')
sns.countplot(x="Year", hue="Genre", data=data)
plt.show()
```

```python
#5 - Crie uma visualização que mostre a média de orçamento de filmes por gênero em cada ano. Em que ano os filmes de ação tiveram maior média de orçamento? E em que ano os filmes de animação tiveram a maior média de orçamento?

plt.figure(figsize=(18, 5))
plt.title("Orçamento médio por gênero em cada ano")
sns.barplot(x="Year",  y="Budget_M", hue="Genre", ci = None, data=data)
plt.show()
```
