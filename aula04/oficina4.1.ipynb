```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set(style="darkgrid")
```

```python
titanic = sns.load_dataset('titanic')
titanic.head()
```

```python
#2 - Crie uma visualizao que mostre a distribuio do valor da passagem (fare), de acordo com a cidade na qual o passageiro embarcou ("embark_town"). Voc pode escolher a tcnica que julgar mais adequada

plt.figure(figsize=(5,5))
plt.title("Distribuio entre valor da passagem e a cidade de embarque")
sns.stripplot(x="embark_town", y="fare", palette=["blue", "red", "orange"], data=titanic, jitter=.20)
plt.show()
```

```python
#3 - Crie uma visualizao que mostre as distribuies do valor da passagem, de acordo com a cidade de embarque, sexo e se o passageiro sobreviveu ou no. Voc pode escolher a tcnica que julgar mais adequada. Que diferena voc consegue notar nas distribuies dos passageiros que sobreviveram em relao aos que no sobreviveram?

sns.catplot(x="embark_town", y="fare", data=titanic, jitter=.20, hue="sex", col="survived", dodge=True)
plt.show()
```
