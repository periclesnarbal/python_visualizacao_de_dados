```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set(style="darkgrid")
```

```python
uri = "https://d11s0xq1vqg039.cloudfront.net/videos_conteudo/visual/micro07/oficina/arquivos/dadosoficina07.txt"
vendas = pd.read_csv(uri)
vendas.head()
```

```python
#2 - crie uma visualizao que mostre a srie temporal das vendas ao longo dos meses.

plt.figure(figsize=(20,5))
plt.title("Vendas de shampoo ao longo do tempo")
sns.lineplot(x="Month", y="Sales", data=vendas, )
plt.show()
```
