```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set(style="darkgrid")
```

```python
df = sns.load_dataset('mpg')
df.head()
```

```python
# 2 - Crie uma visualização que mostre a relação entre a aceleração (‘acceleration’) e o consumo de combustível (‘mpg’) dos veículos, em que seja possível diferenciar o país de origem (‘origin’).

sns.lmplot(x='acceleration', y='mpg', hue='origin', data=df, truncate=False, height=7, legend_out=False)
ax = plt.gca()
ax.set_title('Relação entre acelaração e consumo de combustivel')
plt.show()
```

```python
# 3 - Crie uma visualização que mostre a relação entre a aceleração (‘acceleration’) e o consumo de combustível (‘mpg’) dos veículos, em que seja possível visualizar o ano do modelo (model_year).

sns.lmplot(x='acceleration', y='mpg', hue='model_year', col="model_year", data=df, col_wrap=4, height=2, truncate=False)
plt.show()
```
