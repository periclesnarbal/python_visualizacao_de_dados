```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set(style="darkgrid")
```

```python
df = sns.load_dataset('mpg')
df.head()
```

```python
# 2 - Crie uma visualizao que mostre a relao entre a acelerao (acceleration) e o consumo de combustvel (mpg) dos veculos, em que seja possvel diferenciar o pas de origem (origin).

sns.lmplot(x='acceleration', y='mpg', hue='origin', data=df, truncate=False, height=7, legend_out=False)
ax = plt.gca()
ax.set_title('Relao entre acelarao e consumo de combustivel')
plt.show()
```

```python
# 3 - Crie uma visualizao que mostre a relao entre a acelerao (acceleration) e o consumo de combustvel (mpg) dos veculos, em que seja possvel visualizar o ano do modelo (model_year).

sns.lmplot(x='acceleration', y='mpg', hue='model_year', col="model_year", data=df, col_wrap=4, height=2, truncate=False)
plt.show()
```
