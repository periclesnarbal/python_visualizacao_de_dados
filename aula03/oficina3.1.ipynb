```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv("houses_to_rent_v3.txt")
df.head()
```

```python
#2 - Crie uma visualização que mostre a distribuição dos valores de aluguel, ‘rent amount (R$)’, dos imóveis. Pode utilizar o gráfico que você julgar mais adequado. Qual faixa de valores de aluguel é mais comum?

#https://medium.com/@rachel.hwung/how-to-make-elegant-graphs-quickly-in-python-using-matplotlib-and-seaborn-252344bc7fcf

fig, ax = plt.subplots(figsize=[10,5],frameon = False)
ax.set_title('Distribuição dos valores de aluguel')
sns.distplot(df['rent amount (R$)'], ax=ax)
ax.set_xticks(np.arange(0, 25000 + 1, 2500))
plt.show()
```

```python
#3 - Crie uma visualização que mostre a distribuição dos valores de seguro contra incêndio, 'fire insurance (R$)', dos imóveis. Existem valores incomuns?

fig, ax = plt.subplots(figsize=[10,5],frameon = False)
ax.set_title('Distribuição dos valores de seguro contra incêndio')
sns.distplot(df['fire insurance (R$)'], ax=ax,)
ax.set_xticks(np.arange(0, 300 + 1, 20))
plt.show()
```
