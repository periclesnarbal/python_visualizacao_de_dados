```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv("houses_to_rent_v3.txt")
df.head()
```

```python
#2 - Crie uma visualiza巽達o que mostre a distribui巽達o dos valores de aluguel, rent amount (R$), dos im坦veis. Pode utilizar o gr叩fico que voc棚 julgar mais adequado. Qual faixa de valores de aluguel 辿 mais comum?

#https://medium.com/@rachel.hwung/how-to-make-elegant-graphs-quickly-in-python-using-matplotlib-and-seaborn-252344bc7fcf

fig, ax = plt.subplots(figsize=[10,5],frameon = False)
ax.set_title('Distribui巽達o dos valores de aluguel')
sns.distplot(df['rent amount (R$)'], ax=ax)
ax.set_xticks(np.arange(0, 25000 + 1, 2500))
plt.show()
```
