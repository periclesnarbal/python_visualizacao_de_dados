```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set(style="darkgrid")
```

```python
room = pd.read_csv("room_data.txt")
room.head()
```

```python
plt.figure(figsize=(10,5))
plt.title("Vaariação de temperatura ao longo do tempo")
sns.scatterplot(x="date", y="Temperature", data=room)
plt.show()
```

```python
room['date'] = pd.to_datetime(room['date'])
plt.figure(figsize=(10,5))
plt.title("Vaariação de temperatura ao longo do tempo")
sns.lineplot(x="date", y="Temperature", data=room)
plt.show()
```

```python
hiv = pd.read_csv("hiv_data.txt")
hiv.head()
```

```python
plt.figure(figsize=(10,5))
plt.title('Variação da taxa média de HIV global')
sns.lineplot(x="date", y="HIV Rate", data=hiv)
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.title('Variação da taxa média de HIV global por renda')
sns.lineplot(x="date", y="HIV Rate", hue="Income Level", data=hiv, ci=None)
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.title('Variação da taxa média de HIV global por país')
sns.lineplot(x="date", y="HIV Rate", hue="country", data=hiv, legend=False, ci=None)
plt.show()
```
